```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- (1) If you havent yet, rename the speleothem column for consistency ---
df_d18O_raw.rename(columns={'-d18O': 'd18O'}, inplace=True)

# --- (2) Helper to do interp  crop  residual in one go ---
def interp_and_residual(df_raw, df_filt, raw_col, filt_col, time_col='age'):
    x_raw  = df_raw[time_col].values
    x_filt = df_filt[time_col].values
    y_filt = df_filt[filt_col].values
    
    # linear interp, < min or > max  NaN
    y_interp = np.interp(x_raw,
                         x_filt,
                         y_filt,
                         left=np.nan,
                         right=np.nan)
    
    # attach to a copy, drop the NaNs, compute residual
    df = df_raw[[time_col, raw_col]].copy()
    df['filt_interp'] = y_interp
    df = df.dropna(subset=['filt_interp']).reset_index(drop=True)
    df['residual']  = df[raw_col] - df['filt_interp']
    return df

# --- (3) Run for each pair ---
df_ch4     = interp_and_residual(df_ch4_raw,     df_filt_ch4_low,    raw_col='ch4',    filt_col='filt_ch4')
df_spele   = interp_and_residual(df_d18O_raw,     df_filt_d18O_low,   raw_col='d18O',   filt_col='filt_d18O')
df_fuji    = interp_and_residual(df_fuji_raw,     df_filt_fuji_d18O_low, raw_col='d18O', filt_col='filt_d18O')
df_benthic = interp_and_residual(df_d18O_B_raw,   df_filt_d18O_B_low, raw_col='d18O',   filt_col='d18O_low')

# --- (4) Plot residuals ---
for df, title, ylabel in [
    (df_ch4,     'CH Residual',               'CH (ppb)'),
    (df_spele,   'Speleothem O Residual',   'O ()'),
    (df_fuji,    'Fuji O Residual',         'O ()'),
    (df_benthic, 'Benthic O Residual',      'O ()'),
]:
    plt.figure(figsize=(10, 4))
    plt.plot(df['age'], df['residual'], marker='o', linestyle='-', markersize=2)
    plt.title(title)
    plt.xlabel('Age (ka)')
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def count_state_transitions(age, resid, window=20_000, step=1_000):
    """
    Count how many times the residual series crosses its local median
    within a sliding window.

    Parameters
    ----------
    age   : 1D array of times (must be in ascending or descending order)
    resid : 1D array of residuals (same length as age)
    window: total width of the sliding window (in same units as age)
    step  : step between window centers

    Returns
    -------
    centres        : 1D array of windowcentre times
    n_transitions  : 1D array of counts of mediancrossing flips per window
    """
    # Ensure ascending age (so we can invert axis later consistently)
    if np.any(np.diff(age) < 0):
        age = age[::-1]
        resid = resid[::-1]

    half_w = window / 2
    t_min, t_max = age.min(), age.max()
    centres = np.arange(t_min + half_w, t_max - half_w + step, step)

    n_transitions = []
    for c in centres:
        lo, hi = c - half_w, c + half_w
        mask   = (age >= lo) & (age < hi)
        seg_r  = resid[mask]
        if seg_r.size < 2:
            n_transitions.append(np.nan)
            continue

        med    = np.nanmedian(seg_r)
        states = (seg_r >= med).astype(int)
        flips  = np.diff(states) != 0
        n_transitions.append(flips.sum())

    return centres, np.array(n_transitions)


```

```python

datasets = [
    (df_ch4['age'].values,     df_ch4['residual'].values,     'CH',     'CH (ppb)'),
    (df_spele['age'].values,   df_spele['residual'].values,   'Speleothem O', 'O ()'),
    (df_fuji['age'].values,    df_fuji['residual'].values,    'Fuji O',       'O ()'),
    (df_benthic['age'].values, df_benthic['residual'].values, 'Benthic O',    'O ()'),
]

for age, resid, label, ylabel in datasets:
    centres, n_trans = count_state_transitions(age, resid,
                                               window=20_000,
                                               step=1_000)

    plt.figure(figsize=(10, 4))
    plt.plot(centres, n_trans, lw=1.5)
    plt.gca().invert_xaxis()
    plt.xlabel('Age (ka BP)')
    plt.ylabel(f'No. transitions\n({ylabel})')
    plt.title(f'{label}: flips per 20ka window (median split)')
    plt.grid(alpha=0.4)
    plt.tight_layout()
    plt.show()
```

```python
def crop_residuals(df, t_min=70_000, t_max=640_000, time_col='age'):
    """
    Crop a residualseries DataFrame to a given time window.

    Parameters
    ----------
    df       : pandas.DataFrame
               Must contain a time column (default 'age') and a 'residual' column.
    t_min    : float
               Lower bound of the age window (inclusive).
    t_max    : float
               Upper bound of the age window (inclusive).
    time_col : str
               Name of the time column (default 'age').

    Returns
    -------
    pandas.DataFrame
        A new DataFrame containing only rows where t_min <= time_col <= t_max,
        with its index reset.
    """
    mask = (df[time_col] >= t_min) & (df[time_col] <= t_max)
    return df.loc[mask].reset_index(drop=True)

import numpy as np
import pandas as pd
from scipy.interpolate import interp1d

# 1) Read & prep your precession + obliquity series
pre_path = r"D:\VScode\bipolar_seesaw_CCM\inso_data\pre_800_inter100.txt"
obl_path = r"D:\VScode\bipolar_seesaw_CCM\inso_data\obl_800_inter100.txt"

df_pre_raw = pd.read_csv(pre_path, sep=r'\s+', header=None, engine='python')
df_obl_raw = pd.read_csv(obl_path, sep=r'\s+', header=None, engine='python')

# convert to years BP & ensure increasing age
for df in (df_pre_raw, df_obl_raw):
    df.iloc[:,0] = df.iloc[:,0].abs() * 1000
    df[:] = df.iloc[::-1].reset_index(drop=True)

df_pre_raw.columns = ['age','pre']
df_obl_raw.columns = ['age','obl']


# 2) Function to sample precession & obliquity onto any cropped 'age'
def sample_forcing(df_crop, df_pre, df_obl, age_col='age'):
    """
    Given a df_crop with df_crop[age_col],  
    interpolate df_pre['pre'] and df_obl['obl'] onto those ages.
    Raises ValueError if any age is outside the forcing data's range.
    Returns a new DataFrame with added 'pre' and 'obl' columns.
    """
    ages = df_crop[age_col].values
    # check bounds
    for name, dff in [('precession', df_pre), ('obliquity', df_obl)]:
        amin, amax = dff['age'].min(), dff['age'].max()
        if ages.min() < amin or ages.max() > amax:
            raise ValueError(
                f"{name.capitalize()} data span [{amin:.0f}, {amax:.0f}] "
                f"does not cover cropped ages [{ages.min():.0f}, {ages.max():.0f}]"
            )
    # build interpolators
    f_pre = interp1d(df_pre['age'], df_pre['pre'], kind='linear', bounds_error=True)
    f_obl = interp1d(df_obl['age'], df_obl['obl'], kind='linear', bounds_error=True)

    out = df_crop.copy()
    out['pre'] = f_pre(ages)
    out['obl'] = f_obl(ages)
    return out


# Create a white noise random series with the samelength as the df_pre_raw
white_noise = np.random.normal(0, 1, len(df_pre_raw))
# create a DataFrame with the white noise
df_noise = pd.DataFrame({
    'age': df_pre_raw['age'],
    'white_noise': white_noise
})


def sample_random_noise(df_crop, df_noise, age_col='age'):
    """
    Given a df_crop with df_crop[age_col],  
    interpolate df_noise['white_noise'] onto those ages.
    Raises ValueError if any age is outside the noise data's range.
    Returns a new DataFrame with added 'white_noise' column.
    """
    ages = df_crop[age_col].values
    # check bounds
    amin, amax = df_noise['age'].min(), df_noise['age'].max()
    if ages.min() < amin or ages.max() > amax:
        raise ValueError(
            f"Noise data span [{amin:.0f}, {amax:.0f}] "
            f"does not cover cropped ages [{ages.min():.0f}, {ages.max():.0f}]"
        )
    
    # build interpolator
    f_noise = interp1d(df_noise['age'], df_noise['white_noise'], kind='linear', bounds_error=True)

    out = df_crop.copy()
    out['white_noise'] = f_noise(ages)
    return out

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import warnings

# Suppress future warnings from sklearn
warnings.filterwarnings("ignore", category=FutureWarning)

# Continuous-to-point TE via logistic regression
def compute_TE_cont2point(X, Y, lag=1):
    # Prepare data
    X_prev = X[:-lag].reshape(-1, 1)
    Y_prev = Y[:-lag].reshape(-1, 1)
    Y_next = Y[lag:]
    
    # Model without X: use large C to approximate no regularization
    # model0 = LogisticRegression(C=1e12, solver='lbfgs', max_iter=1000)
    model0 = LogisticRegression(penalty='none', solver='lbfgs', max_iter=1000)
    model0.fit(Y_prev, Y_next)
    p0 = model0.predict_proba(Y_prev)
    
    # Model with X: likewise
    features = np.hstack((Y_prev, X_prev))
    model1 = LogisticRegression(C=1e12, solver='lbfgs', max_iter=1000)
    model1.fit(features, Y_next)
    p1 = model1.predict_proba(features)
    
    # Compute TE as average log-likelihood difference
    idx = np.arange(len(Y_next))
    logp0 = np.log(p0[idx, Y_next])
    logp1 = np.log(p1[idx, Y_next])
    return np.mean(logp1 - logp0)

# Surrogate test
def surrogate_TE_cont2point(X, Y, n_surr=1000):
    TE_obs = compute_TE_cont2point(X, Y)
    TE_surr = []
    for _ in range(n_surr):
        X_shuf = np.random.permutation(X)
        TE_surr.append(compute_TE_cont2point(X_shuf, Y))
    TE_surr = np.array(TE_surr)
    p_value = np.mean(TE_surr >= TE_obs)
    return TE_obs, p_value, TE_surr
```

```python
df_ch4_crop     = crop_residuals(df_ch4,t_min=1000, t_max=600000)
df_spele_crop   = crop_residuals(df_spele,t_min=1000, t_max=600000)
df_fuji_crop    = crop_residuals(df_fuji,t_min=1000, t_max=600000)
df_benthic_crop = crop_residuals(df_benthic,t_min=1000, t_max=600000)

df_ch4_crop     = sample_forcing(df_ch4_crop,   df_pre_raw, df_obl_raw)
df_spele_crop   = sample_forcing(df_spele_crop, df_pre_raw, df_obl_raw)
df_fuji_crop    = sample_forcing(df_fuji_crop,  df_pre_raw, df_obl_raw)
df_benthic_crop = sample_forcing(df_benthic_crop, df_pre_raw, df_obl_raw)

# print the info of the cropped dataframes
print("\nCropped DataFrame Info:")
print("CH Crop DataFrame Info:")
print(df_ch4_crop.info())
print("\nSpeleothem O Crop DataFrame Info:")
print(df_spele_crop.info())
print("\nFuji O Crop DataFrame Info:")
print(df_fuji_crop.info())
print("\nBenthic O Crop DataFrame Info:")
print(df_benthic_crop.info())
# print the column names of the cropped dataframes

# ---------------------------------------------------------------------
# 0.  Imports  (only widely-available packages)
# ---------------------------------------------------------------------
import numpy as np
import pandas as pd
from scipy.special import digamma
from sklearn.neighbors import KDTree
from tqdm.auto import tqdm    # progress bars for long loops

# >>> If you already have 'npeet' installed you can replace the CMI/TE
# >>> helpers below with `from npeet import entropy_estimators as ee`.
# >>> The hand-rolled versions are included so the script runs out
# >>> of the box on a vanilla Python+SciPy stack.

# ---------------------------------------------------------------------
# 1.  Utility: bin precession + binarise residual
# ---------------------------------------------------------------------
def discretise_pre(pre_series, n_bins=6):
    """
    Equal-frequency (quantile) binning of the continuous precession
    forcing.  
    Returns an int array [0  n_bins-1].
    """
    return pd.qcut(pre_series, n_bins, labels=False, duplicates="drop").to_numpy()

def residual_to_state(residual_series):
    """
    Convert residual to a binary 'cold/ warm' state with the median
    as threshold. 0 = low (cold), 1 = high (warm).
    """
    median = np.median(residual_series)
    return (residual_series > median).astype(int)

# ---------------------------------------------------------------------
# 2.  Build pastpresent triplets on an *irregular* grid
# ---------------------------------------------------------------------
def build_lagged_samples(df, lag_yrs, tol_yrs=200,
                          age_col="age",
                          driver_col="pre",
                          target_state_col="state"):
    """
    Align driver_past (X_{t-lag}), target_past (Y_{t-lag}),
    target_present (Y_t) for an *irregularly spaced* time series.

    A sample is kept only if an observation exists within tol_yrs
    of the required lagged timestamp.
    """
    # Ensure chronological order (ascending age)
    df = df.sort_values(age_col).reset_index(drop=True)

    age   = df[age_col].to_numpy()
    x_val = df[driver_col].to_numpy()
    y_val = df[target_state_col].to_numpy()

    # Result containers
    x_past, y_past, y_now = [], [], []

    # For fast nearest-age lookup use numpy searchsorted on 'age'
    for i, t_now in enumerate(age):
        t_lag = t_now - lag_yrs
        # position where 't_lag' would be inserted
        j = np.searchsorted(age, t_lag)
        # choose neighbour (j or j-1) that is *closest* to t_lag
        cand = []
        if j < len(age):
            cand.append(j)
        if j-1 >= 0:
            cand.append(j-1)

        # pick candidate with minimum |age - t_lag|
        if not cand:
            continue
        j_best = min(cand, key=lambda idx: abs(age[idx] - t_lag))

        # keep only if within tolerance
        if abs(age[j_best] - t_lag) <= tol_yrs:
            x_past.append(x_val[j_best])
            y_past.append(y_val[j_best])
            y_now.append(y_val[i])

    return np.asarray(x_past), np.asarray(y_past), np.asarray(y_now)

# ---------------------------------------------------------------------
# 3.  k-NN (Kraskov-style) estimators    minimal implementation
# ---------------------------------------------------------------------
def _knn_distances(data, k):
    """
    Return distance to the k-th nearest neighbour for every point.
    """
    tree = KDTree(data, metric="chebyshev")        # -norm as in Kraskov
    dists, _ = tree.query(data, k=k+1)             # k+1 because point itself
    return dists[:, -1]                            # distance to kth neighbour

def kraskov_conditional_entropy(joint, cond, k=4):
    """
    H(joint | cond)  using Kraskov et al. 2004 (Algorithm 1).
    joint : 2-D array (N  d_joint)     variables (Y, X, )
    cond  : 2-D array (N  d_cond)      conditioning set (X, )
    """
    n = len(joint)
    eps = 1e-15

    # Distances in full joint space
    d_joint = _knn_distances(np.hstack([joint, cond]), k)
    # Count points within that same distance *only in cond-space*
    tree_cond = KDTree(cond, metric="chebyshev")
    m = (tree_cond.query_radius(cond, r=d_joint - eps, count_only=True) - 1)

    return (digamma(n) - np.mean(digamma(m + 1)) + 
            np.mean(np.log(d_joint + eps)))

def transfer_entropy_knn(x_past, y_past, y_now, k=4):
    """
    TE_{XY} = H(Y_t | Y_{t-})  H(Y_t | Y_{t-}, X_{t-})
    All inputs must be 1-D *discrete* int arrays of the same length.
    """
    # reshape for 2-D KDTree input
    y_t     = y_now.reshape(-1, 1)
    y_past  = y_past.reshape(-1, 1)
    x_past  = x_past.reshape(-1, 1)

    h1 = kraskov_conditional_entropy(y_t, y_past,              k=k)
    h2 = kraskov_conditional_entropy(y_t, np.hstack([y_past,
                                                     x_past]), k=k)
    return h1 - h2          # in nats; divide by log(2) for bits

# ---------------------------------------------------------------------
# 4.  One-shot TE with surrogate z-test
# ---------------------------------------------------------------------
def kernel_te_irregular(df, lag_yrs=21000, tol_yrs=200,
                        pre_bins=6, k=4,
                        n_surr=999, random_state=0):
    """
    Compute TE(pre  residual_state) at a given geological lag
    on *irregular* data + permutation (circular shift) surrogate test.
    """
    rng = np.random.default_rng(random_state)

    # 4.1  discretise columns
    df = df.copy()
    df["pre_disc"] = discretise_pre(df["pre"], n_bins=pre_bins)
    df["state"]    = residual_to_state(df["residual"])

    # 4.2  build aligned samples
    x_past, y_past, y_now = build_lagged_samples(
        df,
        lag_yrs   = lag_yrs,
        tol_yrs   = tol_yrs,
        driver_col= "pre_disc",
        target_state_col="state"
    )

    if len(y_now) < 50:
        raise RuntimeError("Too few aligned samples (<50). "
                           "Relax lag/tol or extend time window.")

    # 4.3  observed TE
    te_obs = transfer_entropy_knn(x_past, y_past, y_now, k=k)

    # 4.4  surrogates: circularly shift the driver series
    te_surr = np.empty(n_surr)
    for r in range(n_surr):
        shift = rng.integers(1, len(df)-1)
        df["pre_disc_surr"] = np.roll(df["pre_disc"].to_numpy(), shift)

        xs, ys, yn = build_lagged_samples(
            df,
            lag_yrs   = lag_yrs,
            tol_yrs   = tol_yrs,
            driver_col= "pre_disc_surr",
            target_state_col="state"
        )
        te_surr[r] = transfer_entropy_knn(xs, ys, yn, k=k)

    # 4.5  p-value (one-sided, higher TE means stronger drive)
    p_val = (np.sum(te_surr >= te_obs) + 1) / (n_surr + 1)

    return te_obs / np.log(2), p_val     # return TE in *bits*

# ---------------------------------------------------------------------
# 5.  Example: run on all four proxies
# ---------------------------------------------------------------------
if __name__ == "__main__":
    # Choose parameters once (modify as needed)
    LAG_YRS   = 21000        # canonical precession half-period
    TOL_YRS   = 300          #  tolerance for age matching
    PRE_BINS  = 6
    KNN_K     = 4
    NSURR     = 499

    proxy_dfs = {
        "ch4"     : df_ch4_crop,
        "spele"   : df_spele_crop,
        "fuji"    : df_fuji_crop,
        "benthic" : df_benthic_crop,
    }

    results = {}
    for name, df in proxy_dfs.items():
        te, p = kernel_te_irregular(df,
                                    lag_yrs = LAG_YRS,
                                    tol_yrs = TOL_YRS,
                                    pre_bins= PRE_BINS,
                                    k       = KNN_K,
                                    n_surr  = NSURR)
        results[name] = (te, p)
        print(f"{name:<8}  TE = {te:5.3f} bits   p = {p:6.4f}")

    # >>>  interpret: p < 0.05 (after FDR)  robust information flow
```

```python
# 
#   Kernel / k-NN transfer-entropy on irregular data using PyInform  
# 
import numpy as np
import pandas as pd
from pyinform import transfer_entropy          #  existing package
from tqdm.auto import tqdm                     # nice progress bars

# 
# 1.  Discretisation helpers
# 
def discretise_pre(series, n_bins=6):
    """Equal-frequency bins [0  n_bins-1] for continuous precession."""
    return pd.qcut(series, q=n_bins, labels=False,
                   duplicates="drop").astype(int).to_numpy()

def residual_to_state(series):
    """Binary cold/ warm state using the median as threshold."""
    median = np.median(series)
    return (series > median).astype(int)

# 
# 2.  Align pastpresent triplets on an irregular grid
# 
def build_lagged_samples(df, lag_yrs, tol_yrs=300,
                          age="age", driver="x", target_state="y"):
    """
    Return aligned (x_past, y_past, y_now) arrays where past is
    df[age]  age_now  lag_yrs within tol_yrs.
    """
    df = df.sort_values(age).reset_index(drop=True)

    age_arr   = df[age].to_numpy()
    x_arr     = df[driver].to_numpy()
    y_arr     = df[target_state].to_numpy()

    x_p, y_p, y_n = [], [], []
    for i, t_now in enumerate(age_arr):
        t_lag = t_now - lag_yrs
        j = np.searchsorted(age_arr, t_lag)

        # choose neighbour (j or j-1) closer to t_lag
        idx_cand = [j] if j < len(age_arr) else []
        if j - 1 >= 0:
            idx_cand.append(j - 1)
        if not idx_cand:
            continue
        j_best = min(idx_cand, key=lambda k: abs(age_arr[k] - t_lag))

        if abs(age_arr[j_best] - t_lag) <= tol_yrs:
            x_p.append(x_arr[j_best])
            y_p.append(y_arr[j_best])
            y_n.append(y_arr[i])

    return np.asarray(x_p), np.asarray(y_p), np.asarray(y_n)

# 
# 3.  Transfer-entropy wrapper with surrogate test
# 
def te_irregular(df, lag_yrs=21000, tol_yrs=300,
                 pre_bins=6, n_surr=999, random_state=0):
    """
    TE(pre_disc  residual_state) in *bits* + permutation p-value.
    """
    rng = np.random.default_rng(random_state)

    # 3-A  discretise both series
    df = df.copy()
    df["pre_disc"] = discretise_pre(df["pre"], n_bins=pre_bins)
    df["state"]    = residual_to_state(df["residual"])

    # 3-B  align pastpresent
    x_past, y_past, y_now = build_lagged_samples(
        df, lag_yrs, tol_yrs,
        driver="pre_disc", target_state="state"
    )
    if len(y_now) < 50:
        raise ValueError("Too few aligned samples (<50). Adjust lag/tol.")

    # 3-C  observed TE (k=1 is standard for binary targets)
    te_obs = transfer_entropy(x_past.tolist(),
                              y_now.tolist(),
                              k=1, local=False)

    # 3-D  permutation surrogates (circular shift of driver)
    te_surr = np.empty(n_surr)
    for r in range(n_surr):
        shift = rng.integers(1, len(df) - 1)
        df["pre_disc_surr"] = np.roll(df["pre_disc"].to_numpy(), shift)
        xs, ys, yn = build_lagged_samples(
            df, lag_yrs, tol_yrs,
            driver="pre_disc_surr", target_state="state"
        )
        te_surr[r] = transfer_entropy(xs.tolist(),
                                      yn.tolist(),
                                      k=1, local=False)

    # one-sided p (large TE  stronger drive)
    p_val = (np.sum(te_surr >= te_obs) + 1) / (n_surr + 1)

    return te_obs, p_val             # bits, p-value

# 
# 4.  Run on all four proxies (example)
# 
LAG_YRS  = 21_000   #  half a precession cycle
TOL_YRS  = 300
NSURR    = 499      #  increase for publication-grade tests

proxy_dfs = {
    "CH"      : df_ch4_crop,
    "Spele"    : df_spele_crop,
    "Fuji"     : df_fuji_crop,
    "Benthic"  : df_benthic_crop,
}

results = {}
for name, df in tqdm(proxy_dfs.items()):
    te, p = te_irregular(df,
                         lag_yrs=LAG_YRS,
                         tol_yrs=TOL_YRS,
                         n_surr=NSURR)
    results[name] = (te, p)
    print(f"{name:<8}   TE = {te:6.3f} bits   p = {p:7.4f}")
```
