```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- (1) If you haven’t yet, rename the speleothem column for consistency ---
df_d18O_raw.rename(columns={'-d18O': 'd18O'}, inplace=True)

# --- (2) Helper to do interp → crop → residual in one go ---
def interp_and_residual(df_raw, df_filt, raw_col, filt_col, time_col='age'):
    x_raw  = df_raw[time_col].values
    x_filt = df_filt[time_col].values
    y_filt = df_filt[filt_col].values
    
    # linear interp, < min or > max → NaN
    y_interp = np.interp(x_raw,
                         x_filt,
                         y_filt,
                         left=np.nan,
                         right=np.nan)
    
    # attach to a copy, drop the NaNs, compute residual
    df = df_raw[[time_col, raw_col]].copy()
    df['filt_interp'] = y_interp
    df = df.dropna(subset=['filt_interp']).reset_index(drop=True)
    df['residual']  = df[raw_col] - df['filt_interp']
    return df

# --- (3) Run for each pair ---
df_ch4     = interp_and_residual(df_ch4_raw,     df_filt_ch4_low,    raw_col='ch4',    filt_col='filt_ch4')
df_spele   = interp_and_residual(df_d18O_raw,     df_filt_d18O_low,   raw_col='d18O',   filt_col='filt_d18O')
df_fuji    = interp_and_residual(df_fuji_raw,     df_filt_fuji_d18O_low, raw_col='d18O', filt_col='filt_d18O')
df_benthic = interp_and_residual(df_d18O_B_raw,   df_filt_d18O_B_low, raw_col='d18O',   filt_col='d18O_low')

# --- (4) Plot residuals ---
for df, title, ylabel in [
    (df_ch4,     'CH₄ Residual',               'CH₄ (ppb)'),
    (df_spele,   'Speleothem δ¹⁸O Residual',   'δ¹⁸O (‰)'),
    (df_fuji,    'Fuji δ¹⁸O Residual',         'δ¹⁸O (‰)'),
    (df_benthic, 'Benthic δ¹⁸O Residual',      'δ¹⁸O (‰)'),
]:
    plt.figure(figsize=(10, 4))
    plt.plot(df['age'], df['residual'], marker='o', linestyle='-', markersize=2)
    plt.title(title)
    plt.xlabel('Age (ka)')
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def count_state_transitions(age, resid, window=20_000, step=1_000):
    """
    Count how many times the residual series crosses its local median
    within a sliding window.

    Parameters
    ----------
    age   : 1D array of times (must be in ascending or descending order)
    resid : 1D array of residuals (same length as age)
    window: total width of the sliding window (in same units as age)
    step  : step between window centers

    Returns
    -------
    centres        : 1D array of window‐centre times
    n_transitions  : 1D array of counts of median‐crossing flips per window
    """
    # Ensure ascending age (so we can invert axis later consistently)
    if np.any(np.diff(age) < 0):
        age = age[::-1]
        resid = resid[::-1]

    half_w = window / 2
    t_min, t_max = age.min(), age.max()
    centres = np.arange(t_min + half_w, t_max - half_w + step, step)

    n_transitions = []
    for c in centres:
        lo, hi = c - half_w, c + half_w
        mask   = (age >= lo) & (age < hi)
        seg_r  = resid[mask]
        if seg_r.size < 2:
            n_transitions.append(np.nan)
            continue

        med    = np.nanmedian(seg_r)
        states = (seg_r >= med).astype(int)
        flips  = np.diff(states) != 0
        n_transitions.append(flips.sum())

    return centres, np.array(n_transitions)


```

```python
def crop_residuals(df, t_min=70_000, t_max=640_000, time_col='age'):
    """
    Crop a residual‐series DataFrame to a given time window.

    Parameters
    ----------
    df       : pandas.DataFrame
               Must contain a time column (default 'age') and a 'residual' column.
    t_min    : float
               Lower bound of the age window (inclusive).
    t_max    : float
               Upper bound of the age window (inclusive).
    time_col : str
               Name of the time column (default 'age').

    Returns
    -------
    pandas.DataFrame
        A new DataFrame containing only rows where t_min <= time_col <= t_max,
        with its index reset.
    """
    mask = (df[time_col] >= t_min) & (df[time_col] <= t_max)
    return df.loc[mask].reset_index(drop=True)

import numpy as np
import pandas as pd
from scipy.interpolate import interp1d

# 1) Read & prep your precession + obliquity series
pre_path = r"D:\VScode\bipolar_seesaw_CCM\inso_data\pre_800_inter100.txt"
obl_path = r"D:\VScode\bipolar_seesaw_CCM\inso_data\obl_800_inter100.txt"

df_pre_raw = pd.read_csv(pre_path, sep=r'\s+', header=None, engine='python')
df_obl_raw = pd.read_csv(obl_path, sep=r'\s+', header=None, engine='python')

# convert to years BP & ensure increasing age
for df in (df_pre_raw, df_obl_raw):
    df.iloc[:,0] = df.iloc[:,0].abs() * 1000
    df[:] = df.iloc[::-1].reset_index(drop=True)

df_pre_raw.columns = ['age','pre']
df_obl_raw.columns = ['age','obl']


# 2) Function to sample precession & obliquity onto any cropped 'age'
def sample_forcing(df_crop, df_pre, df_obl, age_col='age'):
    """
    Given a df_crop with df_crop[age_col],  
    interpolate df_pre['pre'] and df_obl['obl'] onto those ages.
    Raises ValueError if any age is outside the forcing data's range.
    Returns a new DataFrame with added 'pre' and 'obl' columns.
    """
    ages = df_crop[age_col].values
    # check bounds
    for name, dff in [('precession', df_pre), ('obliquity', df_obl)]:
        amin, amax = dff['age'].min(), dff['age'].max()
        if ages.min() < amin or ages.max() > amax:
            raise ValueError(
                f"{name.capitalize()} data span [{amin:.0f}, {amax:.0f}] "
                f"does not cover cropped ages [{ages.min():.0f}, {ages.max():.0f}]"
            )
    # build interpolators
    f_pre = interp1d(df_pre['age'], df_pre['pre'], kind='linear', bounds_error=True)
    f_obl = interp1d(df_obl['age'], df_obl['obl'], kind='linear', bounds_error=True)

    out = df_crop.copy()
    out['pre'] = f_pre(ages)
    out['obl'] = f_obl(ages)
    return out


```

```python
# simulate two independent white-noise signals
import numpy as np
from pyinform import transfer_entropy
x = np.random.randn(10000)
y = np.random.randn(10000)

def discretize(a, bins):
    edges = np.histogram_bin_edges(a, bins=bins)
    return np.digitize(a, edges) - 1

for B in [2, 5, 10, 20]:
    xd = discretize(x, B)
    yd = discretize(y, B)
    te = transfer_entropy(xd, yd, k=1)
    print(f"{B:2d} bins → TE ≈ {te:.4f}")
```
